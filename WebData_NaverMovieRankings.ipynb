### 자동화를 위한 코드

'''
https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20220703
'''
- 날짜만 변경하면 우리가 원하는 기간 만큼 데이터 얻을 수 있음

```python
date = pd.date_range("2022.01.01", periods=100, freq="D") # 1월1일부터 100일 하루(Day) 단위로
date
```

```python
date[0].strftime("%Y-%m-%d") # 형태 변경
```

```python
# 문자열 format
test_string = "Hi, I'm {name}"
test_string.format(name="Jungsu")
```

```python
import time
from tqdm import tqdm

movie_date = []
movie_name = []
movie_point = []

for today in tqdm(date):
    url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date={date}"
    response = urlopen(url.format(date=today.strftime("%Y%m%d")))
    soup = BeautifulSoup(response, "html.parser")
    
    end = len(soup.find_all("td", "point"))
    movie_date.extend([today for _ in range(end)])
    movie_name.extend([soup.select("div.tit5")[n].find("a").text for n in range(end)])
    movie_point.extend([soup.find_all("td", "point")[n].string for n in range(end)])
    
    time.sleep(0.5)
```

```python
len(movie_date), len(movie_name), len(movie_point)
```

```python
movie = pd.DataFrame({
    "date": movie_date,
    "title": movie_name,
    "ratings": movie_point
})

movie.tail()
```

```python
movie.info()
```

```python
movie["ratings"] = movie["ratings"].astype(float)
movie.info()
```

```python
# 데이터 저장
movie.to_csv('naver_movie_data.csv', sep=",", encoding="utf-8")
```
