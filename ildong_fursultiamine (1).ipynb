```python
from selenium.webdriver.common.keys import Keys

driver.get('https://biz.some.co.kr/insightSocialTransition')
driver.find_element_by_xpath('//*[@id="search_keyword"]').clear()
driver.find_element_by_xpath('//*[@id="search_keyword"]').send_keys('푸르설티아민')
driver.find_element_by_xpath('//*[@id="include_keywords"]').clear()
driver.find_element_by_xpath('//*[@id="include_keywords"]').send_keys('아로나민')
# driver.find_element_by_xpath('//*[@id="startDate"]').click()
driver.find_element_by_xpath('//*[@id="startDate"]').clear()
driver.find_element_by_xpath('//*[@id="startDate"]').send_keys('2020/01/01')

time.sleep(1)

driver.find_element_by_css_selector('#ui-datepicker-div > div.ui-datepicker-buttonpane.ui-widget-content > button.ui-datepicker-close.ui-state-default.ui-priority-primary.ui-corner-all').click()

time.sleep(4)


# driver.find_element_by_css_selector('#execute_analysis').click()
try: 
    driver.find_element_by_xpath('//*[@id="execute_analysis"]').click()
except:
    button = driver.find_element_by_xpath("//*[@id='execute_analysis']")
    driver.execute_script("//*[@id='execute_analysis']", button)

driver.execute_script("window.scrollTo(300, 600)")

time.sleep(10)

driver.find_element_by_css_selector("#documentDiv > ul > li:nth-child(4) > a").click()
# driver.find_elements_by_css_selector("#documentDiv > ul > li.on > a").click()
# driver.find_element_by_css_selector('#documentDiv > ul > li:nth-child(4) > a').click()
# driver.find_element_by_css_selector('#searchOptionsBox > div > div:nth-child(4) > div.search-options-inputs > div > div > span:nth-child(7) > label').click()
# driver.find_element_by_css_selector('#documentDiv > ul > li.on > a').send_keys(Keys.ENTER)
time.sleep(5)
html = driver.page_source
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, "html.parser")
print(soup.prettify())
```

```python
list_soup = soup.find_all("dd", "original-contents-text")

f = open('f_report.txt', 'w', encoding='utf-8')

for item in list_soup:
    f.write(item.get_text())
f.close()
```
